```python
# Import lớp Image
from PIL import Image
from imgtools import *
```

```python
# Đường dẫn ảnh
my_path = 'D:/hoctap/hk7/tgmt/image/i03.jpg'
```

```python
# Đọc ảnh 
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (100, 100, 125, 180)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Xác định vị trí cần dán
paste_position = (950, 150)
img.paste(cropped_image, paste_position)
display(img)
```
